```python
from randomuser import RandomUser
```

```python
emails = []
f_names = []
l_names = []
passwords = []

for i in range(0,3):

  user_list = RandomUser.generate_users(5000,{'nat': 'us'})

  for u in user_list:
    e = u.get_email()
    if e not in emails:
      emails.append(e)
      f_names.append(u.get_first_name())
      l_names.append(u.get_last_name())
      passwords.append(u.get_password())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'FirstName':f_names,'LastName':l_names,'Email':emails,'Password':passwords})
```

```python
df
```

```python
len(pd.unique(df['Email']))
```

```python
df.to_csv('drive/My Drive/customer_data.csv')
```

Making room table

```python
room_num = []
room_type = []
room_status = []
for i in range(0,500):
  room_num.append(i)
  room_status.append('Available')
  if i>=0 and i<100:
    room_type.append('Single Bedroom')
  elif i>=100 and i<200:
    room_type.append('Double Bedroom')
  elif i>=200 and i<300:
    room_type.append('Suite')
  elif i>=300 and i<400:
    room_type.append('Executive Suite')
  elif i>=400 and i<500:
    room_type.append('President Suite')


```

```python
room_df = pd.DataFrame({'Room_number':room_num,'Type':room_type,"Status":room_status})
```

```python
room_df
```

```python
room_df['Type'].value_counts()
```

```python
room_df.to_csv('drive/My Drive/hotelroom_data.csv',index=False)
```

Making staff table

```python
import re
```

```python
staff_emails = ['ayesha@hotel.com','zoha@hotel.com','nida@hotel.com']
staff_f_names = ['Ayesha','Zoha','Nida']
staff_l_names = ['Masood','Hayat','Tanveer']
staff_passwords = ['ash123','zoha123','nida123']
role = ['Owner','Manager','Receptionist']
shift = ['am','am','am']


staff_list = RandomUser.generate_users(100,{'nat': 'ca'})

for i in range(0,len(staff_list)):

  u = staff_list[i]

  f = u.get_first_name()
  l = u.get_last_name()
  p = u.get_password()
  
  e = u.get_email()
  em = re.sub('example', 'hotel',e)
  em = re.sub('-', '_', em)
  
  if em not in staff_emails:
    staff_emails.append(em)
    staff_f_names.append(f)
    staff_l_names.append(l)
    staff_passwords.append(p)


    if i>=0 and i<25:
      role.append('Housekeeping')
    elif i>=25 and i<50:
      role.append('Waiter')
    elif i>=50 and i<60:
      role.append('Valet')
    elif i>=60 and i<75:
      role.append('Concierge')
    else:
      role.append('Security')

    if i<50:
      shift.append('am')
    else:
      shift.append('pm')

    
```

```python
staff_df = pd.DataFrame({'FirstName':staff_f_names,'LastName':staff_l_names,'Type':role,'Email':staff_emails,'Password':staff_passwords,'Shift':shift})
```

```python
staff_df
```

```python
staff_df.to_csv('drive/My Drive/staff_data.csv')
```
